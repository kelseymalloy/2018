I am particularly interest in the precipitation that occured June 2, 2017 over and south of the Gulf States. Was the specific humidity and meridional wind at 850 mb correlated at all? I don't really expect a correlation, but this is good practice for domain-related correlations, interpolation to certain levels, etc.

```python
import matplotlib as mpl
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display, Math, Latex
from numpy.random import rand
from pylab import rcParams, pcolor, show,colorbar, xticks, yticks, title, figure
from IPython.display import Image
import netCDF4 as nc
```

# Consider some (pairs of) numbers: a scatterplot

```python
# Define a figure: scatterplot. 

mpl.rcParams.update(mpl.rcParamsDefault)
%matplotlib inline
rcParams['figure.figsize'] = (11, 6.5)

def fig5():
    fig,ax=plt.subplots(1,1)
    ax.plot(v850_domain,q850_domain,'.')
    ax.set_xlabel('850 mb Meridional Wind [m s-1]');ax.set_ylabel('850 mb Specific Humidity [g kg-1]')
    ax.set_title(utc)
    
```

```python
#Data reading:

# read in netCDF
filename = "MERRA2_400.inst3_3d_asm_Nv.20170602.nc4"
f=nc.Dataset(filename, mode='r')
```

```python
## Extract variables from the file object
print(f.variables.keys())
time = f.variables['time'][:] # minutes since 1:30 UTC
lon = f.variables['lon'][:] # longitude
lat = f.variables['lat'][:] # latitude
lev = f.variables['lev'][:] # level
#print(lon)

p0 = f.variables['PS'][:][:][:][:] / 100 # surface pressure
delp = f.variables['DELP'][:][:][:][:] / 100 # press difference between layers
```

```python
# LOOP THROUGH ALL TIMES IN FILE
tcount=0 # initialize
while tcount<1: # just going to look at one time step for now
    #convert levels into pressure levels (look at MERRA documentation, but levels are desginated as sigma, so no pressure array exists yet)
    x=0
    p_start=np.zeros((len(time),len(lev),len(lat),len(lon)),dtype='float')
    p_start[:,:,:,:]=0.01  # top pressure for every x-y point
    pressure=p_start # initialize pressure array

    while x<len(lev):# go through every level
        if x==0:  # first level is already set at .01 
            x=x+1
        else:
            pressure[tcount][x][:][:] = pressure[tcount][x-1][:][:] + delp[tcount][x][:][:]  # fill in pressure array with pressure from previous index plus delta pressure
            x=x+1

            
    # read in variables of interest          
    v = f.variables['V'][tcount][:][:][:] # v-component of wind
    q = f.variables['QV'][tcount][:][:][:] # mass fraction of cloud liquid water

    #interpolate data to 850 mb
    pressure=pressure[tcount,:,:,:] # which time step?
    v850=np.zeros((len(lat),len(lon)))
    q850=np.zeros((len(lat),len(lon)))
    upper_level=850 # which level do you wanna see? 
    x=0
    while x<len(lon): # go through every longitude
        y=0 # intitialize for every new lon
        while y<len(lat): # go through every latitude
            z=0 # initialize for every new lat
            while z<len(lev): # go though every level
                if (pressure[z,y,x] >= upper_level): # picks index immed. > 850mb
                    #print(pressure[z,y,x])
                    v850[y,x]=v[z,y,x]+((upper_level-pressure[z,y,x])*((v[z-1,y,x]-v[z,y,x])/(pressure[z-1,y,x]-pressure[z,y,x]))) # interpolate to 300mb between indices with values below and above 300mb
                    #print(v850[y,x])
                    q850[y,x]=q[z,y,x]+((upper_level-pressure[z,y,x])*((q[z-1,y,x]-q[z,y,x])/(pressure[z-1,y,x]-pressure[z,y,x])))
                    
                    z=len(lev) # get out of loop for levels or will overwrite
                else:
                    z=z+1
            y=y+1
        x=x+1
  
    tcount=tcount+1
    
#print(v850)
```

```python
# Only extract variables in particular domain
start_lat = 28 #N
end_lat = 34 #N
start_lon = -100 #W
end_lon = -83 #W

# find indices with preferred lat and lon coordinates
i_start = int((np.abs(lat-start_lat)).argmin())
print(i_start)
#i_start = i_start[0]
i_end = int((np.abs(lat-end_lat)).argmin())
print(i_end)
#i_end = i_end[0]
j_start = int((np.abs(lon-start_lon)).argmin())
print(j_start)
#j_start = j_start[0]
j_end = int((np.abs(lon-end_lon)).argmin())
print(j_end)
#j_end = j_end[0]

# only want values in this domain
v850_domain = v850[i_start:i_end,j_start:j_end]
q850_domain = q850[i_start:i_end,j_start:j_end] * 100 # g kg-1


utc=130+(tcount*300) # set UTC time
utc=format(utc,'04d') # format the string for title and output file



# PLOT IT
fig5()
```

### Beyond the scatterplot: histogram methods
The dots cover each other, so a better way to look at this is to _contour or image the data density in bins_,
counting how many points fall in each bin. Let's do it first in the general plotting package Matplotlib, then in a more specifically statistal graphics package called Seaborn.


```python
q850_domain_fixedlat = q850_domain[-1,:] 
junk = plt.hist(q850_domain_fixedlat)
```

There seem to be 3 "modes" of specific humidity values, but may be better to look at animation of moving bins.

```python
from numpy import histogram as nphistogram
from JSAnimation.IPython_display import display_animation, anim_to_html
from matplotlib import animation as animation


def getHistBinNumAni(data,totalframes=None,showpts=True):
    #ion()
    if totalframes is None:
        totalframes = min(len(data)-1,100)
    fig = figure()
    ax = fig.gca()

    n, bins, patches = plt.hist(data, totalframes, normed=1, facecolor='green', alpha=0.0)
    if showpts:
        junk = plt.plot(data,0.2*np.ones_like(data),'bo')
    def animate(i):
        n, bins = nphistogram(data, i+1, normed=False)
        #print n
        ax.set_ylim(0,1.1*n.max())
        for j in range(len(n)):
            rect,h = patches[j],n[j]
            #print h.max()
            x = bins[j]
            w = bins[j+1] - x
            rect.set_height(h)
            rect.set_x(x)
            rect.set_width(w)
            rect.set_alpha(0.75)
        #fig.canvas.draw()
    
    ani = animation.FuncAnimation(fig, animate, totalframes, repeat=False)
    return ani
```

```python
ani = getHistBinNumAni(q850_domain_fixedlat)
display_animation(ani, default_mode='once')
```

10-12 bins seems to best represent the tri-modal nature of the 850-mb specific humidity at this fixed latitude.

```python
x = v850_domain[-1,:];x=x-np.mean(x)
y = q850_domain[-1,:];y=y-np.mean(y)
r = np.sum(x*y)/np.std(x,ddof=1)/np.std(y,ddof=1)/(len(v850_domain[-1,:])-1)
print (r)
F = np.arctanh(r)
Fp95 = F+1.96/np.sqrt(len(v850_domain[-1,:])-3)
Fm95 = F-1.96/np.sqrt(len(v850_domain[-1,:])-3)
rp=np.tanh(Fp95)
rm=np.tanh(Fm95)
print ("r= %1.3f (%1.3f,%1.3f)"%(r,rm,rp))
```

The r (correlation coefficient value is low (< 0.5); therefore, the meridional wind and specific humidity at 850 mb in this domain and timeframe are not well-correlated.
